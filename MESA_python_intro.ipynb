# Reading data from multiple files

```python
file_list = ['file0.csv',
'file1.csv',
'file2.csv',
'file3.csv',
'file4.csv',
'file5.csv',
'file6.csv',
'file7.csv',
'file8.csv',
'file9.csv']
```

```python
#out data structure is going to be a list of lists
data = []

for input_file in file_list:
    file_data = [[],[]]
    file = open(directory+input_file, 'r')
    for row in file.readlines():
        row = row.strip()
        row = row.split(',')
        file_data[0].append(row[0])
        file_data[1].append(row[1])
    data.append(file_data)        
```

```python
print(data[0])
```

```python
#this is where we do the acual plotting
figure(1, figsize(14, 6))
for item in data:
    plot(item[0],item[1], linestyle='-', marker='o') 
    #plot(item[0],item[1], linestyle='', marker='.') 
    #want to plot a horizontal line at the average of each point?
    axhline(y=average([float(j) for j in item[1]]), linestyle='--', linewidth=5) #color=(0.0,0.0,0.0)
       
xlabel("X Axis", fontsize = 20)
ylabel("Y Axis", fontsize = 20) 
xlim(-0.1,10.1)
ylim(0.,5001)
#save the following plot as an image file
savefig('folder_of_files/jackson_pollock.png', bbox_inches='tight')
```
